```python
 #Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data by creating data frame
df = pd.read_csv('/Users/andrewmelcer/Desktop/movies.csv')

```

```python
# Preview that data was uploaded correctly by previewing it with head function 
df.head()
```

```python
# Check to see if there is missing data with for loop

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# Check data types for each column

df.dtypes
```

```python
# USE fillna function in order to fill in missing values 
df = df.fillna(0)
```

```python
#change data type for budget and gross to get rid of decimal point

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Order values by revenue

df.sort_values(by=['gross'], inplace =False, ascending=False)
```

```python
# Change display options to view all rows of data frame

pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# Budget high correlation
# Company high correlation 

# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.show()
```

```python
# Add more detail to scatter plot

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Plot budget vs gross using seaborn package

sns.regplot(x='budget', y ='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Begin looking at correlation 

df.corr()
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show
```

```python
# Look at different film companies 

df.head()
```

```python
# Look at different film companies 

# Convert and assign numbers to company names

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
        
```

```python
df.head()
```

```python
#Use correlation_matrix on df_numerized 

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show
```

```python
#Easier way to see the correlaion between value 

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort the pairs using sort values function

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Show sorted pairs with a high correlation value 

high_correlation = sorted_pairs[(sorted_pairs) > 0.5]

high_correlation
```
